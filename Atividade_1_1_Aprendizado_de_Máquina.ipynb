Para cada exemplo abaixo, indique se o problema é:

i. predição ou inferência

ii. regressão, classificação, ou agrupamento

iii. supervisionado ou não-supervisionado

iv. quais os valores de `n` (número de exemplos) e `p` (número de atributos)

Justifique cada resposta.


a) Coletamos os dados de 1000 empresas e, para cada uma delas, coletamos dados de receita, lucro, número de empregados, indústria, crescimento e salário do presidente. 

Queremos criar um modelo para obter mais informações sobre quais dos fatores coletados influenciam o salário do presidente.

**Resposta:**

*i.* Inferência - queremos saber quais atributos influenciam na resposta.

*ii.* Regressão - pois o salário do presidente é um valor numérico. (poderia ser uma classificação caso o salário fosse transformado em valor categórico, como crescimento ou decrescimento).

*iii.* Supervisionado - pois temos os dados da resposta (crescimento do salário do presidente) para os 1000 exemplos (empresas)

*iv.* n=1000, p=4 (receita, lucro, número de empregados e indústria)


b) Uma empresa quer lançar um novo produto e deseja saber se o produto será um sucesso ou não.

Eles coletaram dados de 20 produtos similares que foram lançados anteriormente, coletando se ele foi um sucesso ou fracasso, o preço cobrado, gasto com propaganda, preço dos concorrentes e outras 10 variáveis.

**Resposta:**

*i.* Predição - pois queremos prever o resultado.

*ii.* Classificação - pois a variável resposta é categórica, e, nesse caso, binária (sim ou não).

*iii.* Supervisionado - pois temos dados da variável resposta para os 20 produtos.

*iv.* n=20 (produtos similares); p=13 (preço cobrado, gasto, preço concorrentes, outras 10)


c) Queremos identificar se um paciente está propenso a ter problemas cardiovasculares a partir de um grupo de indicadores de risco. 

Coletamos os dados de 1000 pacientes, com a informação se eles tiveram ou não problemas cardiovasculares, e com os seguintes indicadores:
- Idade
- Altura
- Peso
- Sexo
- Triglicerídeos
- Pressão arterial (máximo)
- Pressão arterial (mínimo)
- Colesterol
- Nível de atividade física
- Frequência de uso de tabaco 
- Frequência de abuso de bebidas álcoolicas



**Resposta:**

*i.* Predição - estamos interessados na resposta.

*ii.* Classificação - variável resposta sim ou não.

*iii.* Supervisionado - temos os dados da variável resposta para cada um dos exemplos.

*iv.* n=1000 (pacientes); p=11 (indicadores)


d) Uma rede de supermecados deseja identificar diferentes perfis de consumidores para realizar uma campanha comercial. Para tal, ela possui os seguintes dados de 10.000 consumidores, obtidos por meio de seu programa de recompensas:

**Resposta:**

*i.* Predição - Problema de identificar perfil, e não necessariamente das variáveis que influenciam no perfil.

*ii.* Agrupamento - Não temos variável resposta, apenas o agrupamento dos exemplos em um perfil.

*iii.* Não supervisionado - Não temos variável resposta.

*iv.* n=10000 (consumidores); p=7

### Exercício 1.3 - [1 ponto]

Veremos agora sobre modos de avaliar o desempenho de um modelo. Suponha que você treinou um modelo de regressão e um de classificação sobre um conjunto de dados de treino. Ao realizar os testes, obteve os seguintes resultados, onde Y é o valor alvo, e Modelo é a saída do seu modelo:

Regressão:

Observação |  Y (alvo)  | Saída do Modelo
-----------|------------|-------------------
1          |   3.0     | 4.0
2          |   5.0     | 7.0
3          |   1.0     | 4.0
4          |   -2.0    | -5.0
5          |   10.0    | 5.0

Classificação:

Observação |  Y (alvo)  | Saída do Modelo
-----------| -------|-------------------
1          |   azul     | vermelho
2          |   azul     | azul
3          |   vermelho | azul
4          |   azul     | azul
5          |   vermelho | vermelho







a) Qual o erro (MSE) do modelo no caso da regressão para cada entrada?

Indique quais as contas realizadas. Exemplo: $\sqrt{2^2 + 3^2} = 4$

```python
# Regressão
y = [
    3.0 ,
    5.0 ,
    1.0 ,
    -2.0,
    10.0,
]
y_pred = [
    4.0,
    7.0,
    4.0,
    -5.0,
    5.0,
]
def mse(y, y_pred):
    return sum([(yi-yi_pred)**2 for yi, yi_pred in zip(y,y_pred)])/len(y)
mse(y, y_pred)
```

b) Qual o erro (Taxa de acerto) do modelo no caso da classificação para cada entrada?

Indique quais as contas realizadas.

```python
#Classificação
y = [
    "azul",
    "azul",
    "vermelho",
    "azul",
    "vermelho",
]
y_pred = [
    "vermelho",
    "azul",
    "azul",
    "azul",
    "vermelho",
]
def accuracy(y, y_pred):
    return sum([yi == yi_pred for yi, yi_pred in zip(y,y_pred)])/len(y)
accuracy(y, y_pred)
```

```python
import pandas as pd
dataset_values = np.hstack([dataset.data, dataset.target.reshape(-1,1)])
dataset_columns = dataset.feature_names+dataset.target_names

df = pd.DataFrame(dataset_values, columns = dataset_columns)
```

```python
df.sample(3, random_state=1)
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize = (15,20))
ax = fig.gca()
histogram = df.hist(ax = ax)
```

```python
dataset_meta_info = [
    'median income in block group',
    'median house age in block group',
    'average number of rooms per household',
    'average number of bedrooms per household',
    'block group population',
    'average number of household members',
    'block group latitude',
    'block group longitude',
    'median house value for California districts, expressed in hundreds of thousands of dollars ($100,000).'
]
cols_bins = [
    8, #ok
    10, #ok
    list(range(0,44,4)),       #14,
    list(range(0,16,2)),       #14,
    list(range(0,11000,1000)), #20,
    list(range(0,16,2)),       #20,
    12, 
    12,
]
```

```python
figsize = (22,14)
rows = 2
cols = 4

axs = plt.figure(figsize = figsize).subplots(rows, cols)
axs = axs.flat
for ax, col_name, meta_info, bins in zip(
    axs,dataset_columns[:-1], dataset_meta_info[:-1], cols_bins
):
    df.hist(column = col_name, ax = ax, bins = bins)
    ax.set_ylabel('# instances')
    ax.set_xlabel(meta_info)
    if isinstance(bins, list):
        ax.set_xticks(bins)
```
